```python
#上辺寸法
number1 = 10
#下辺寸法
number2 = 20
#高さ寸法
number3 = 5

print("次の台形の面積を求めよ。")
print(f"上辺は{number1}cmで下辺は{number2}cm、高さは{number3}㎝とする。")

print(f"式 ({number1}＋{number2})×{number3}÷2")

#台形の面積の計算
number4 = number1 + number2
number5 = number4 * number3 /2

print(f"答え　台形の面積は　{number5}ｃｍ²です。")
```
