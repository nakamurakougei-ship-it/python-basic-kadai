```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    print(f"{self.name}さんは{self.age}才")

    if self.age > 20:
      print("大人です")

    else:
      print("大人ではありません")

hito1 = Human("侍太郎", 35)
hito2 = Human("侍二郎", 15)
hito3 = Human("侍三郎", 25)
hito4 = Human("侍四郎", 10)
hito5 = Human("侍五郎", 5)
hito6 = Human("侍花子", 45)

member = [hito1, hito2, hito3, hito4, hito5, hito6]

for hito in member:
  hito.check_adult()
```
